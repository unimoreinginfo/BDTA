1. Crea il DataFrame **vendite** che contiene la quantità di torte, biscotti e taralli (colonne) nei diversi giorni della settimana (righe).

  Inserisci valori casuali compresi tra 5 e 50 (np.random.seed(100)).

```python
import pandas as pd
import numpy as np

np.random.seed(100)

vendite=pd.DataFrame(np.random.randint(5,50,size=(7,3)),
                     columns=['Torte','Biscotti','Taralli'],
                     index=['Lunedi','Martedi','Mercoledi','Giovedi','Venerdi','Sabato','Domenica'])
vendite
```

2. Visualizza le vendite di torte durante tutta la settimana.

```python
vendite['Torte']
```

3. Ottieni le vendite del giovedi.

```python
vendite.loc['Giovedi',:]
```

4. Visualizza le vendite di biscotti e taralli dal mercoledì alla domenica.

```python
vendite.loc['Mercoledi':'Domenica','Biscotti':'Taralli']
```

5. Qual è il totale delle vendite per ogni prodotto durante la settimana?

```python
vendite_biscotti=vendite['Biscotti'].sum()
vendite_torte=vendite['Torte'].sum()
vendite_taralli=vendite['Taralli'].sum()

print(vendite_torte, vendite_biscotti, vendite_taralli)
```

6. Qual è il giorno con il massimo delle vendite di torte?

```python
vendite['Torte'].idxmax()
```

7. Aggiungi la colonna **Totale Vendite Giorno** che rappresenta la somma delle vendite di tutti i prodotti per ciascun giorno.

```python
vendite_totali_giornaliere = pd.Series([vendite.loc['Lunedi',].sum(), vendite.loc['Martedi',].sum(), vendite.loc['Mercoledi',].sum(), vendite.loc['Giovedi',].sum(), vendite.loc['Venerdi',].sum(), vendite.loc['Sabato',].sum(), vendite.loc['Domenica',].sum()], index=['Lunedi','Martedi','Mercoledi','Giovedi','Venerdi','Sabato','Domenica'])
vendite_totali_giornaliere
#vendite['Totale'] = vendite.sum(axis=1) si può fare così oppure
vendite['Totale'] = vendite_totali_giornaliere
vendite
```

8. Quali giorni hanno avuto un totale di vendite giornaliere superiori a 30?

```python
vendite['Totale'] > 30
```
